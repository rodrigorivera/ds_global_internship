```python
# Authors - Captain Capucine and her Trusty side kick Ulisse 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk

# Here we import the cryptocurrency data into
# two distinct files. One variable is called train_dataset
# that will be used to train our machine learning algorithm.
# The second variable called test_dataset will be used to test
# our trained model, to see if our model is accurate or not.


# Ulisse - Capucine or whoever views this if you have a question
# about any code I have written ask me plus make a highlight in the code.


train_dataset = pd.read_csv('01_train_70pct_crypto_1583_2013-2019.csv')
test_dataset = pd.read_csv('02_test_30pct_crypto_1583_2013-2019.csv')


print(train_dataset)
print(test_dataset)


# Below down to ^Above^ I am attempting to find correlations between variables.
# Doing so will allow us to seperate the unimportant data from the important data.
# If we can find the variables that have casual relationships with each other,
# when we build our predictive model, our model will be more accurate. We want to use
# causal relationships between variables to build a accurate model. We want to remove
# non-casual relationships from our data. I tried to plot a visualization to give a better
# understanding of whats going on, however my visualisation sucks, so next time I work on this
# I am going to look for a better way to demonstrate the degree of casualness between variables.

correlations_between_variables = train_dataset.corr()
figure = plt.figure()
ax = figure.add_subplot(111)
cax = ax.matshow(correlations_between_variables, cmap='coolwarm',vmin=-1, vmax=1)
figure.colorbar(cax)
ticks = np.arange(0, len(train_dataset.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(train_dataset.columns)
ax.set_yticklabels(train_dataset.columns)
plt.show()

# ^Above^ 


```
